`pd.cut` is a function in pandas that allows you to segment and sort data into bins. It is particularly useful for dividing continuous data into discrete intervals. It returns a categorical object or a DataFrame with bins as categories.

### Syntax

```python
pd.cut(x, bins, right=True, labels=None, retbins=False, precision=3, include_lowest=False, duplicates='raise')
```

### Parameters
- **`x`**: Array-like, the data to be binned.
- **`bins`**: 
  - Integer (defines the number of bins to create), or
  - Sequence (defines the bin edges explicitly).
- **`right`**: If `True`, bins are right-inclusive (default is `True`).
- **`labels`**: Labels for the resulting bins. If `False`, integer indicators for bins are returned.
- **`retbins`**: If `True`, it returns the bins along with the results.
- **`precision`**: Precision for bin edges when they are numeric.
- **`include_lowest`**: If `True`, the first interval includes the lowest value.
- **`duplicates`**: If bin edges are not unique, raises an error unless set to `'drop'`.

### Returns
- A pandas Categorical object representing the bin each element falls into.
- Optionally, the actual bin edges if `retbins=True`.

---

### Examples

#### 1. Divide data into equal-width bins
```python
import pandas as pd

data = [1, 7, 5, 4, 6, 3, 9, 8, 2, 0]
binned = pd.cut(data, bins=3)
print(binned)
```
Output:
```
[(0.0, 3.0], (6.0, 9.0], (3.0, 6.0], (3.0, 6.0], (6.0, 9.0], (0.0, 3.0], (6.0, 9.0], (6.0, 9.0], (0.0, 3.0], (0.0, 3.0]]
```

#### 2. Custom bin edges
```python
bins = [0, 3, 6, 9]
binned = pd.cut(data, bins=bins, labels=["Low", "Medium", "High"])
print(binned)
```
Output:
```
[Low, High, Medium, Medium, High, Low, High, High, Low, Low]
Categories (3, object): ['Low' < 'Medium' < 'High']
```

#### 3. Include lowest value
```python
binned = pd.cut(data, bins=[0, 3, 6, 9], include_lowest=True)
print(binned)
```
Output:
```
[(0.0, 3.0], (6.0, 9.0], (3.0, 6.0], (3.0, 6.0], (6.0, 9.0], [0.0, 3.0], (6.0, 9.0], (6.0, 9.0], (0.0, 3.0], [0.0, 3.0]]
```

#### 4. Return bin edges
```python
binned, bins = pd.cut(data, bins=3, retbins=True)
print(binned)
print(bins)
```
Output:
```
[(0.0, 3.0], (6.0, 9.0], (3.0, 6.0], (3.0, 6.0], (6.0, 9.0], (0.0, 3.0], (6.0, 9.0], (6.0, 9.0], (0.0, 3.0], (0.0, 3.0]]
[0. 3. 6. 9.]
```

## Perform clustering based on gower distance 

````python
import gower
#gow_distances = gower.gower_matrix(car_df)
````

````python
# Calculate distance based on both, numerical and categorical variables
gow_distances = gower.gower_matrix(car_df, cat_features=categ_bool)
````

````python
gow_distances
````

````python
# Perfrom UMAP dimensionality reduction
umap_obj = umap.UMAP(n_neighbors=30, metric='precomputed')
embedding_mtx = umap_obj.fit_transform(gow_distances)

uplot.points(umap_obj)
````

````python
umap.plot.connectivity(umap_obj, show_points=True)
````

````python
umap.plot.diagnostic(umap_obj, diagnostic_type='pca')
````

````python
# Create dendrogram based on precomputed distance
gow_distances_condens = squareform(gow_distances)

linkage_matrix = linkage(
    gow_distances_condens, 
    method='average'
)
````

````python
cophenetic_corr(
    linkage_matrix=linkage_matrix,
    data_mtx = gow_distances,
    distance_measure='precomputed'
)
````

````python
# Plot the dendrogram
plt.rcParams['figure.figsize'] = [12, 30] 
_ = dendrogram(
    linkage_matrix, 
    orientation='left', 
    labels=car_df.index,
    leaf_font_size=7
)
````

````python
# Plot the dendrogram
plt.rcParams['figure.figsize'] = [12, 30] 
_ = dendrogram(
    linkage_matrix, 
    orientation='left', 
    labels=car_sec_info['price_bins'].to_numpy(),
    leaf_font_size=7
)
````

````python
# Calculate inconsistence matrix
depth=4
incosistent_mtx = inconsistent(linkage_matrix,depth)

# Plot inconsistency scores for given depth
plt.figure(figsize=(10, 7))
plt.plot(incosistent_mtx[:,-1])
plt.scatter(range(incosistent_mtx.shape[0]),incosistent_mtx[:,-1])
plt.title('Inconsistency values at depth = {}'.format(depth))
plt.xlabel('Index')
plt.ylabel('Inconsistency coefficient')
plt.show()
````

````python
# Perform clustering based on inconsistency
# and print clusters
clusters = fcluster(Z=linkage_matrix, t=1.6, criterion='inconsistent', R=incosistent_mtx)
clusters
````

````python
labels = [
    'clust {} - price bin {}'.format(clust, price) 
    for clust, price in zip(clusters, car_sec_info['price_bins'].to_numpy())
]

dendr_colors = plot_cluster_dendrogram(
    linkage_matrix=linkage_matrix,
    dataset_df=car_df,
    clusters=clusters,
    leaf_font_size=7,
    labels=labels
)
````

````python
_ = plot_silouethes_agglomer(
    data_df=gow_distances,
    clusters=clusters,
    dendr_colors=dendr_colors,
    embedding_mtx=embedding_mtx,
    distance_measure='precomputed',
    figsize=(16,12)
)
````

````python
car_sec_info['cluster'] = clusters
clust_size = car_sec_info['cluster'].value_counts().sort_index(ascending=False)
clust_size
````
