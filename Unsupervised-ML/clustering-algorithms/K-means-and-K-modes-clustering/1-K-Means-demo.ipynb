CASE 2: k-means++ initialization


````python
#Defining a list containing initial cluster centers
center_list = []

#Select first center
random_idx = np.random.choice(data_point_idxs, size=1)[0]
center_list.append(x[random_idx])

#Calculate distances from the first center
distance = np.linalg.norm(x - center_list[0], axis=1)
distance_mtx = np.expand_dims(distance, axis=1)

#First iter mark
next_center = 1
````

Initialization of cluster centers, likely for the **K-Means++** clustering algorithm, where the first center is chosen randomly, and subsequent centers are chosen based on the distances from existing centers.


### Explanation of Each Step:

1. **Define `center_list`**:
   ```python
   center_list = []
   ```
   This list will hold the coordinates of the selected cluster centers.

2. **Select the First Center**:
   ```python
   random_idx = np.random.choice(data_point_idxs, size=1)[0]
   center_list.append(x[random_idx])
   ```
   - A random index (`random_idx`) is selected from the available data point indices (`data_point_idxs`).
   - The data point at this index in `x` is chosen as the first center and added to `center_list`.

3. **Calculate Distances from the First Center**:
   ```python
   distance = np.linalg.norm(x - center_list[0], axis=1)
   ```
   - Compute the Euclidean distance between each data point in `x` and the first cluster center (`center_list[0]`).
   - `np.linalg.norm(x - center_list[0], axis=1)` computes the norm (distance) row-wise along axis 1.

4. **Expand the Distance Matrix**:
   ```python
   distance_mtx = np.expand_dims(distance, axis=1)
   ```
   - Reshape the distance array into a column vector. This ensures the matrix can be used for further calculations in subsequent iterations.

5. **Mark the Next Center**:
   ```python
   next_center = 1
   ```
   - Indicates that the algorithm is preparing to select the next center.

---


````python
#Generate centers

while len(center_list) < NUM_CLUSTER:
    
    #For each point, find distance from closest center
    dist_from_closest = np.min(distance_mtx, axis=1)
    #Transform to probabilities
    selection_probs = dist_from_closest/dist_from_closest.sum()
    
    #select new center based on this probabilities
    center_pt_idx = np.random.choice(data_point_idxs, p=selection_probs)
    
    #Assign new center dict
    center_list.append(x[center_pt_idx])
    next_center+=1
    
    #Calculate distance from the new center 
    distance = np.linalg.norm(x - x[center_pt_idx], axis=1)
    distance_mtx = np.column_stack([distance_mtx, distance])
    
clust_centers = np.array(center_list)
````

The **K-Means++ initialization** process for selecting initial cluster centers:

---


#### 1. **Define Loop Condition**
```python
while len(center_list) < NUM_CLUSTER:
```
- The loop runs until the required number of cluster centers (`NUM_CLUSTER`) is selected.
- `center_list` keeps track of the cluster centers selected so far.

---

#### 2. **Find Distance from Closest Center**
```python
dist_from_closest = np.min(distance_mtx, axis=1)
```
- For each data point, calculate its distance from the **nearest center**.
- `np.min(distance_mtx, axis=1)` finds the smallest distance for each point across all currently selected centers.

---

#### 3. **Transform Distances to Probabilities**
```python
selection_probs = dist_from_closest / dist_from_closest.sum()
```
- The distances are normalized to create probabilities:
  - Points farther from existing centers are assigned higher probabilities.
  - Points closer to existing centers are assigned lower probabilities.
- This ensures diversity in center selection.

---

#### 4. **Select a New Center**
```python
center_pt_idx = np.random.choice(data_point_idxs, p=selection_probs)
```
- A new center is selected randomly, weighted by the probabilities in `selection_probs`.

---

#### 5. **Add the New Center to the List**
```python
center_list.append(x[center_pt_idx])
next_center += 1
```
- The selected data point is added to `center_list`, and the counter `next_center` is incremented.

---

#### 6. **Update Distance Matrix**
```python
distance = np.linalg.norm(x - x[center_pt_idx], axis=1)
distance_mtx = np.column_stack([distance_mtx, distance])
```
- Compute distances from the newly selected center to all points in `x`.
- Append these distances as a new column in `distance_mtx`.

---

#### 7. **Store Final Cluster Centers**
```python
clust_centers = np.array(center_list)
```
- Convert the list of centers (`center_list`) into a NumPy array for use in further computations.

---

### Key Points:
- This implementation ensures that initial cluster centers are spaced apart, reducing the likelihood of poor clustering results.
- It's part of the **K-Means++ initialization** strategy, which significantly improves K-Means clustering performance compared to random initialization.

---


````python
#Plot the k-means++ initial cluster centers

plt.figure(figsize=(7, 6))
sns.scatterplot(blobs_df, x='Feature 1', y='Feature 2')
color_list = ['green', 'orange', 'blue']
for clust_id, center in enumerate(clust_centers):
    plt.scatter(center[0], center[1], marker='X', s=200, color=color_list[clust_id], label=clust_id)
    
plt.title('K-means ++ initialized centroids')
plt.legend()
````

K-Means Clustering

````python
# Set MAX_ITER AND TOLERANCE
MAX_ITER = 100
TOLERANCE = 1e-4

#Variable for current iteration
curr_iter = 0

#Iterate until convergence 
#Or until MAX_ITER is reached

while True:
    
    #Each element of this list is np array containing distances from all points
    #to the certain center
    distance_list = []
    
    #List that will be used to update centers
    new_centers = []
    
    #Iterate through the centers
    for center in clust_centers:
        
        #Calculate distance from all the points to the current center and add to the list
        distance_list.append(np.linalg.norm(x - center, axis=1))
        
    #Stack elements of distance list to 2D array
    distance_np = np.column_stack(distance_list)
    #Find index of row containing minimum element
    point_cluster = np.argmin(distance_np, axis=1)
    
    #Update centers
    for cluster_id in cluster_ids:
        new_centers.append(x[point_cluster==cluster_id].mean(axis=0))
    #TRansform new centers to np array
    new_centers = np.array(new_centers)
    
    #Check if loop should be stopped due to MAX_ITER
    if curr_iter >= MAX_ITER:
        break
    
    #Check if centers have moved 
    center_difference = np.linalg.norm(clust_centers - new_centers, axis = 1)
    
    #Check if centers are stable 
    if center_difference.max() < TOLERANCE:
        break 
    #Make new centers clust_centers
    clust_centers = new_centers
    
    #Increase iteration counter 
    curr_iter += 1
    
````

````python
#Plot the final result
plt.figure(figsize=(6, 5))

sns.scatterplot(blobs_df, x='Feature 1', y='Feature 2', hue=point_cluster, palette=color_list, alpha=0.4)

for clust_id, center in enumerate(clust_centers):
    plt.scatter(center[0], center[1], marker='X', s=200, color=color_list[clust_id])
    
plt.title('Final clustering')
````

INERTIA BASED

````python
# Initialize inertia to 0 
inertia = 0 

#Loop goign through data points for each cluster
for clust_id in cluster_ids:
    
    #Calculate inertia portion for current cluster 
    clust_points = x[point_cluster == clust_id]
    distance_from_center = np.linalg.norm(clust_points - clust_centers[clust_id], axis=1)
    
    #Increase inertia
    inertia += np.sum(distance_from_center**2)
````
