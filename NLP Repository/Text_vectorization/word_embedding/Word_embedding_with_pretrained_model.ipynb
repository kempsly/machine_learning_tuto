```python
import gzip
import shutil

# Path to the compressed file
input_gz_file = '/content/GoogleNews-vectors-negative300.bin.gz'

# Path for the output decompressed file
output_bin_file = '/content/GoogleNews-vectors-negative300.bin'

# Open the compressed file in binary mode
with gzip.open(input_gz_file, 'rb') as f_in:
    # Open the output file in binary mode
    with open(output_bin_file, 'wb') as f_out:
        # Copy data from compressed file to output file
        shutil.copyfileobj(f_in, f_out)

print("Decompression completed successfully.")
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize  # Assuming you use NLTK for tokenization

# Example corpus
corpus = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?",
]

# Tokenize the corpus into sentences
tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

# Train Word2Vec model
model = Word2Vec(tokenized_corpus, vector_size=100, window=5, min_count=1, sg=0)

# Access word vectors
vector = model.wv['document']  # Example: get vector for the word 'document'
print(vector)
```

#Similarity score between to words

```python
matrix = wv.vectors
print("shape of matrix is",matrix.shape)
```

```python
w1 = wv.key_to_index['bread']
w2 = wv.key_to_index['butter']

v1 = matrix[w1,:]
v2 = matrix[w2,:]

print(f"Norm of bread is {np.linalg.norm(v1)}")
print(f"Norm of butter is {np.linalg.norm(v2)}")
```

```python
cos_sim = (np.dot(v1,v2))/(np.linalg.norm(v1)*np.linalg.norm(v2))
print(f'cosine similarity is {cos_sim}')
```
