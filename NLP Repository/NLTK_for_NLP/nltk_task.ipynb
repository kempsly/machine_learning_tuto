#Stemming with nltk

##Porter stemmer Implementation

```python
from nltk.stem import PorterStemmer
```

```python
porter = PorterStemmer()
words = ['visible', 'view', 'visibly', 'vision']
for word in words:
  print(f"{word} --->{porter.stem(word)}")
```

```python
porter = PorterStemmer()
words = ['generous', 'generate', 'generously', 'generation']
for word in words:
  print(f"{word} --->{porter.stem(word)}")
```

##Snowball stemmer

```python
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer(language='english')
words = ['generous', 'generate', 'generously', 'generation']
for word in words:
  print(f"{word} --->{snowball.stem(word)}")
```

##Lancaster Stemmer

```python
from nltk.stem import LancasterStemmer
lancaster = LancasterStemmer()
words = ['generous', 'generate', 'generously', 'generation']
for word in words:
  print(f"{word} --->{lancaster.stem(word)}")
```

```python
words = ['visible', 'view', 'visibly', 'vision']
for word in words:
  print(f"{word} --->{lancaster.stem(word)}")
```

##Regex Stemmer

```python
from nltk.stem import RegexpStemmer
reg = RegexpStemmer(regexp='ing|s$|ables$',min=4)
words = ['generous', 'generate', 'generously', 'generation']
for word in words:
  print(f"{word} --->{reg.stem(word)}")
```

##Word sense disambiguation

Word sense disambiguation (WSD) is a natural language processing task that involves determining the meaning of a word based on its context in a given sentence or document. The goal of WSD is to identify the correct sense or meaning of a word when it has multiple possible meanings (senses) depending on the context in which it appears.

Here's a concise overview of the concept and how it is typically addressed in NLP:

### Why Word Sense Disambiguation (WSD)?

Words in natural language often have multiple meanings, known as polysemy, which can vary based on the surrounding words or context. For example, consider the word "bank," which can refer to a financial institution or the side of a river. Understanding which meaning is intended in a particular context is crucial for accurate language understanding and downstream NLP tasks like machine translation, information retrieval, and sentiment analysis.

### Techniques for Word Sense Disambiguation:

1. **Dictionary-based Methods:**
   - Use of lexical resources like WordNet to map words to their possible senses.
   - Identify context clues (surrounding words) to disambiguate between senses.

2. **Supervised Machine Learning:**
   - Train a machine learning model using labeled datasets where each word is annotated with its correct sense.
   - Features include word embeddings, context words, part-of-speech tags, etc.

3. **Unsupervised and Semi-supervised Learning:**
   - Clustering algorithms to group similar words based on context.
   - Graph-based algorithms to propagate information across words and senses.

4. **Neural Network Approaches:**
   - Use of neural networks, particularly transformers, to learn contextual representations and predict word senses.
   - Pre-trained language models like BERT, RoBERTa, and GPT have been applied to WSD tasks.

### Steps in Word Sense Disambiguation:

1. **Preprocessing:**
   - Tokenize the input text and identify words that require disambiguation.

2. **Feature Extraction:**
   - Extract features from the context of the ambiguous word, such as neighboring words, part-of-speech tags, and word embeddings.

3. **Model Building:**
   - Choose a suitable WSD model (e.g., supervised classifier, neural network).
   - Train the model on labeled data or leverage pre-trained models.

4. **Prediction:**
   - Apply the trained model to predict the correct sense of ambiguous words in new sentences.

### Challenges in Word Sense Disambiguation:

- **Data Sparsity:** Labeled datasets for WSD can be limited and may not cover all senses of each word.
- **Context Dependency:** The meaning of a word can change significantly based on its surrounding words and phrases.
- **Ambiguity Resolution:** Some words have highly ambiguous senses, making accurate disambiguation challenging.

Word sense disambiguation is a fundamental problem in NLP, and advancements in techniques like deep learning have improved the accuracy and scalability of WSD systems. However, it remains an active area of research due to its complexity and importance in various NLP applications.

```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
```

```python
a1= lesk(word_tokenize('The building has a device to jam the signal'),'jam')
print(a1,a1.definition())
a2 = lesk(word_tokenize('I am stuck in a traffic jam'),'jam')
print(a2,a2.definition())
```

```python
a1= lesk(word_tokenize('what is the current time ?'),'current')
print(a1,a1.definition())
a2 = lesk(word_tokenize('air current'),'current')
print(a2,a2.definition())
```

##NER

Named Entity Recognition (NER) is a natural language processing (NLP) task that involves identifying and classifying named entities (e.g., names of persons, organizations, locations, dates, quantities) within a body of text. The goal of NER is to locate and categorize specific entities mentioned in the text into predefined categories.

### Importance of Named Entity Recognition:

NER plays a crucial role in various NLP applications, including:

- **Information Extraction:** NER helps extract structured information from unstructured text, enabling the retrieval of key entities.
  
- **Question Answering:** Identifying named entities aids in answering questions about specific entities mentioned in the text.
  
- **Entity Linking:** Linking detected entities to knowledge bases or databases enriches the understanding of text and enhances information retrieval.

- **Text Summarization:** Identifying important entities helps summarize text by focusing on key topics or entities.

### Techniques and Approaches for Named Entity Recognition:

1. **Rule-based Systems:**
   - Use hand-crafted rules and patterns to identify named entities based on linguistic features (e.g., capitalization, part-of-speech tags).
  
2. **Statistical and Machine Learning Approaches:**
   - Utilize supervised learning algorithms (e.g., Conditional Random Fields, Hidden Markov Models, neural networks) trained on labeled datasets.
   - Features for training include word embeddings, part-of-speech tags, and contextual information.

3. **Deep Learning Models:**
   - Employ deep learning architectures like Bidirectional LSTMs (Long Short-Term Memory networks), Transformers (e.g., BERT, RoBERTa), and CRFs (Conditional Random Fields) for end-to-end NER.
   - Pre-trained language models fine-tuned on NER datasets achieve state-of-the-art performance.

### Steps in Named Entity Recognition:

1. **Text Preprocessing:**
   - Tokenize the input text and apply necessary preprocessing steps (e.g., stemming, lemmatization).

2. **Feature Extraction:**
   - Extract features like part-of-speech tags, word embeddings, and contextual information around each word.

3. **Model Building:**
   - Choose a suitable NER model (e.g., CRF, BiLSTM-CRF, BERT-NER).
   - Train the model on labeled datasets containing annotated named entities.

4. **Prediction:**
   - Apply the trained model to new text to identify and classify named entities into predefined categories (e.g., PERSON, ORGANIZATION, LOCATION).

### Challenges in Named Entity Recognition:

- **Ambiguity:** Some words can have multiple meanings or categories based on context, making entity classification challenging.
  
- **Rare Entities:** Limited annotated data for rare or less frequent entities can impact model performance.
  
- **Entity Overlapping:** Entities may overlap or nest within each other (e.g., "New York City" as a location).

Named Entity Recognition is a fundamental task in NLP, and advancements in deep learning models have significantly improved its accuracy and scalability. State-of-the-art NER systems leverage large-scale pre-trained language models and fine-tuning techniques to achieve high performance on diverse text corpora.

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
text1 = "Apple is an American company based out of California"
for w in nltk.word_tokenize(text1):
  for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(w))):
     if hasattr(chunk, 'label'):
        print(chunk.label(), ' '.join(c[0] for c in chunk))
```
